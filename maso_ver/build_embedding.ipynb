### 임베딩 매트릭스 생성 

```python
from stemming.porter2 import stem
import re
from konlpy.tag import Mecab
mecab = Mecab()
```

```python
def make_list(fname, l= 'ko'):
    spaced = []
    with open(fname) as f:
        if l == 'ko':
            spaced = [['START', ] + mecab.morphs(i.strip()) + ['END', ] for i in f.readlines()]
        else:
            spaced = [['START', ] + [stem(j) for j in i.lower().split()] + ['END', ] for i in f.readlines()]
    return(spaced)
```

```python
train_ko = make_list('korean-parallel-corpora/korean-english-v1/korean-english-park.train.ko',l='ko')
len(train_ko)
test_ko = make_list('korean-parallel-corpora/korean-english-v1/korean-english-park.test.ko',l='ko')
len(test_ko)


train_en = make_list('korean-parallel-corpora/korean-english-v1/korean-english-park.train.en',l='en')
len(train_en)
test_en = make_list('korean-parallel-corpora/korean-english-v1/korean-english-park.test.en',l='en')
len(test_en)
```

```python
from itertools import zip_longest, chain


tr_en_ko = [ list(filter(None.__ne__, list(chain.from_iterable(list(zip_longest(i, j[1:-1]))))))  for i, j in zip(train_ko, train_en) ]
te_en_ko = [ list(filter(None.__ne__, list(chain.from_iterable(list(zip_longest(i, j[1:-1]))))))  for i, j in zip(test_ko, test_en) ]

full_text = tr_en_ko + te_en_ko
```

```python
with open("embed_train.txt", "wt") as f:
    f.writelines([" ".join(i) + '\n' for i in full_text])
```
